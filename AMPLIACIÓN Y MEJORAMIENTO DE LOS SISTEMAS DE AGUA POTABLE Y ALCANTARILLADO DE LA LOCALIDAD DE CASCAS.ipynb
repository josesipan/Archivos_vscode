<p><b>1.1.2 Contexto y situación problemática</b></p>

<p style="text-align: justify;">La localidad de Cascas alberga una población cuya actividad económica principal es la agricultura y la ganadería, con producción vitivinícola de importancia regional. A pesar de ello, la infraestructura de saneamiento básico presentaba, al momento de la formulación del estudio, una condición de deterioro severo que ponía en riesgo directo la salud pública de sus habitantes.</p>

</div>

<div style="font-family: 'Times New Roman', Times, serif; font-size: 12pt; text-align: justify;">

<p style="text-align: justify;">El sistema de abastecimiento de agua potable operaba con tuberías de asbesto cemento con más de 40 años de antigüedad, sin proceso de cloración ni tratamiento, con un servicio discontinuo de apenas 2 a 3 horas por día, y con pérdidas en la red estimadas en aproximadamente el 70 % del caudal captado. La cobertura de conexiones domiciliarias de agua alcanzaba el 70 % de las viviendas.</p>

<p style="text-align: justify;">La red de alcantarillado sanitario estaba conformada en su mayoría por tuberías de concreto simple normalizado (CSN) Ø8" con una antigüedad superior a 40 años, en estado de colapso parcial. Los emisores descargaban directamente a la quebrada Collacate sin ningún tipo de tratamiento previo. El 35.6 % de la población carecía de conexión domiciliaria de desagüe. Como consecuencia, se registraba un alto índice de morbilidad por enfermedades gastrointestinales y parasitarias de origen hídrico, particularmente en la población infantil.</p>

<p style="text-align: justify;">Ante esta problemática, la Municipalidad Provincial de Gran Chimú formuló el estudio de preinversión a nivel de perfil, inscrito en el Banco de Proyectos del SNIP con código N° 10415, en el marco de la Ley N° 28802 del Sistema Nacional de Inversión Pública y su Reglamento D.S. N° 221-2006-EF. El presente expediente técnico constituye el desarrollo a nivel definitivo de dicho perfil.</p>

</div>

<div style="font-family: 'Times New Roman', Times, serif; font-size: 12pt; text-align: justify;">

<p><b>1.1.3 Justificación del factor de mayoración censal</b></p>

<p>
Se identificó una actividad minera a 5 km de la localidad, cuyo inicio de operaciones se proyecta en un horizonte de 10 años dentro del período de diseño del sistema. En este sentido, se aplica un factor censal de <b>1.05</b> sobre la población registrada en el censo de referencia.
</p>

<p>
La presencia de dicha actividad minera generará los siguientes efectos sobre la demanda hídrica: (i) crecimiento poblacional por migración laboral hacia la zona; (ii) incremento de la actividad comercial y de servicios en Cascas como núcleo urbano de referencia; y (iii) posible presión sobre los recursos hídricos compartidos en la cuenca. La mayoración del 5 % sobre el dato censal refleja de manera conservadora este incremento adicional de demanda que no está recogido en la tasa de crecimiento intercensal histórica, la cual fue calculada con información anterior al descubrimiento del yacimiento.
</p>

<p>
Adicionalmente, todas las longitudes planimétricas del proyecto han sido escaladas aplicando un factor topográfico de <b>1.100</b>, que corrige la planimetría de los planos del estudio base a las condiciones reales del trazado en campo.
</p>

$$

$$

<p><b>1.5.2 Cámara de Rejas</b></p>

<p>
La cámara de rejas es la primera unidad del tren de tratamiento. Retiene sólidos gruesos y flotantes. El diseño se basa en la velocidad de flujo a través de las rejas, que debe estar entre 0.60 y 1.00 m/s para evitar colmatación prematura.
</p>

<p>
Caudal de diseño:
<i>Q</i><sub>md</sub> = 13.76 L/s = 0.01 m<sup>3</sup>/s
</p>

<p>
Separación entre barras:
e = 25 mm (rejas de desbaste)
</p>

<p>
Ancho del canal:
b = 0.30 m (adoptado)
</p>

<p>
Velocidad de flujo:
v = 0.75 m/s (adoptado, dentro del rango 0.60–1.00 m/s)
</p>

<p style="text-align:center;">

$$
A=\frac{Q}{v}
=\frac{0.01}{0.75}
=0.0183\ \mathrm{m^2}
=183.4\ \mathrm{cm^2}
$$

</p>

<p style="text-align:center;">

$$
y=\frac{A}{b}
=\frac{0.0183}{0.30}
=0.061\ \mathrm{m}
=6.1\ \mathrm{cm}
\longrightarrow
\text{Adoptado: }y=0.25\ \mathrm{m}
$$

</p>

<p>
Longitud del canal de rejas:
L = 1.50 m (permite maniobra del rastrillo limpiador)
</p>

<p>
Las rejas se fabricarán en acero inoxidable AISI 304, con perfil de 1/4'' × 1½''. Se incluye by-pass de emergencia y vertedero rectangular de control en la salida.
</p>

<p><b>1.5.3 Tanque Imhoff</b></p>

<p>
El Tanque Imhoff es la unidad de tratamiento primario. Cumple simultáneamente la función de sedimentador de sólidos suspendidos y digestor anaerobio de lodos. El diseño sigue el procedimiento del RNE OS.090 y las normas CEPIS.
</p>

<p><b>Cálculo de la cámara de sedimentación:</b></p>

<p>
Caudal de diseño:
<i>Q</i><sub>md</sub> = 13.76 L/s = 49.52 m<sup>3</sup>/h
</p>

<p>
Período de retención hidráulico:
<i>t</i><sub>r</sub> = 1.5 horas (RNE OS.090)
</p>

<p style="text-align:center;">

$$
A_s = Q_{md}\times t_r
=49.52\times1.5
=74.28\ \mathrm{m^2}
$$

</p>

<p>
Adoptando largo (L) = 11.78 m y ancho (a) = 3.02 m
</p>

<p style="text-align:center;">

$$
A_s=L\times a
=11.78\times3.02
=35.58\ \mathrm{m^2}
\approx35.55\ \mathrm{m^2}\ (\mathrm{OK})
$$

</p>

<p>
Profundidad del sedimentador:
<i>h</i><sub>sed</sub> = 2.00 m (mínimo normativo)
</p>

<p>
Ángulo de la ranura de paso: 45°; abertura de ranura = 0.25 m
</p>

<p><b>Cálculo de la cámara de digestión de lodos:</b></p>

<p>
Volumen de digestión requerido:
</p>

<p style="text-align:center;">

$$
V_{dig}=0.06\times P_f\times\left(\frac{1}{1000}\right)\ \mathrm{m^3/hab}
$$

</p>

<p>

$$
V_{dig}=0.06\times6095\times\left(\frac{1}{1000}\right)
$$

</p>

<p>
No, usando fórmula estándar:
</p>

<p style="text-align:center;">

$$
V_{dig}=
\frac{\text{Tasa de lodo fresco}\times P_f\times\text{Tiempo de digestión}}{1000}
$$

</p>

<p>
Tasa de lodo fresco = 0.1 L/hab/día × 6095 hab = 610 L/día
</p>

<p>
Tiempo de digestión a 18 °C = 40 días (CEPIS)
</p>

<p style="text-align:center;">

$$
V_{dig}
=
\frac{610\times40}{1000}
=
24.4\ \mathrm{m^3}
$$

</p>

<p>
Adoptado: V<sub>dig</sub> = 388 m<sup>3</sup> (incluye factor de seguridad de 1.15)
</p>

Dimensiones totales del Tanque Imhoff:
11.78 m largo × 7.70 m ancho × 8.87 m profundidad total.

=
0.4\ \mathrm{días}
$$

</p>

<p><b>1.5.7 Caseta de cloración</b></p>

<p>
El efluente final pasa por una cámara de contacto de cloro de flujo pistón. Se aplica hipoclorito de calcio al 70 % mediante dosificador de tabletas. El tiempo de contacto mínimo requerido es de 30 minutos (CEPIS). La dosis de cloro residual en el efluente final debe ser ≥ 0.5 mg/L antes de la disposición final.
</p>

<p><b>1.6 Descripción de Riesgos Naturales Asociados al Sistema</b></p>

<p><b>1.6.1 Contexto de amenazas naturales</b></p>

<p>
La localidad de Cascas se ubica en la vertiente occidental de los Andes a 1 279 m.s.n.m., en la zona de transición entre costa y sierra. Esta posición la expone a múltiples amenazas naturales que pueden afectar la infraestructura de saneamiento. El análisis de riesgos identifica cinco amenazas principales y propone medidas de mitigación específicas para cada una.
</p>

<p><b>1.6.2 Sismicidad</b></p>

<p>
El departamento de La Libertad se ubica en la Zona Sísmica 3 según la NTE E.030 (Diseño Sismorresistente, 2018), con aceleración pico del suelo PGA = 0.35g. Los componentes más vulnerables son los reservorios, las captaciones y la PTAR, por ser estructuras de concreto con juntas rígidas.
</p>

<p>
Las medidas de mitigación incluyen:
</p>

<ul style="font-family:'Times New Roman', Times, serif; font-size:12pt;">

<li>Diseño sismorresistente de todas las estructuras conforme a la NTE E.030.</li>

<li>Anclajes adicionales en tuberías con cambio de dirección.</li>

<li>Juntas flexibles en conexiones de tuberías a estructuras de concreto.</li>

<li>Revisión periódica de juntas y sellos tras eventos sísmicos.</li>

</ul>


\<p><b>1.6.3 Huaicos y deslizamientos</b></p>

<p>
Durante eventos extremos asociados al Fenómeno El Niño, las quebradas cercanas a Cascas pueden activarse generando huaicos y deslizamientos que afecten las líneas de conducción y redes enterradas.
</p>

<p>
Las medidas de mitigación contempladas en el proyecto son:
</p>

<ul style="font-family:'Times New Roman', Times, serif; font-size:12pt;">

<li>Enterrado mínimo de 0.80 m en tuberías de conducción.</li>

<li>Protección mediante dados de concreto en cruces de quebradas.</li>

<li>Construcción de muros de contención y disipadores donde el perfil hidráulico lo requiera.</li>

<li>Programa anual de limpieza de cunetas y alcantarillas.</li>

</ul>

<p><b>1.6.4 Sequía y variabilidad hidrológica</b></p>

<p>
El cambio climático podría reducir el caudal de estiaje de los manantiales. Por ello, el diseño incorpora dos fuentes independientes de abastecimiento, permitiendo mantener la continuidad del servicio aun cuando una de ellas presente disminuciones temporales de caudal.
</p>

<p>
El proyecto recomienda implementar un programa permanente de monitoreo hidrológico mediante aforos semestrales y control de niveles piezométricos, con el fin de detectar oportunamente tendencias de reducción en la disponibilidad hídrica y adoptar medidas preventivas de operación.
</p>

<p><b>1.6.5 Riesgo asociado a la actividad minera</b></p>

<p>
El inicio proyectado de actividades mineras en las proximidades de la localidad constituye un riesgo potencial para la sostenibilidad del sistema de abastecimiento de agua potable, debido a la posible alteración de la recarga de los acuíferos que alimentan los manantiales de Los Chimbiles y La Peña, así como por el incremento de la demanda hídrica regional.
</p>

<p>
Con el fin de reducir este riesgo, el proyecto considera las siguientes medidas preventivas:
</p>

<ul style="font-family:'Times New Roman', Times, serif; font-size:12pt;">

<li>Implementar un programa de monitoreo semestral de caudales en ambas fuentes de abastecimiento.</li>

<li>Realizar monitoreos periódicos de la calidad del agua para detectar oportunamente posibles alteraciones fisicoquímicas o microbiológicas.</li>

<li>Coordinar con la Autoridad Nacional del Agua (ANA) el seguimiento de nuevos derechos de uso de agua que puedan afectar la disponibilidad del recurso hídrico en la cuenca.</li>

<li>Actualizar el balance hídrico del sistema cuando la actividad minera entre en operación, verificando que la oferta disponible continúe siendo suficiente para satisfacer la demanda poblacional proyectada.</li>

<li>Establecer un plan de contingencia para la operación del sistema ante escenarios de reducción significativa del caudal de los manantiales.</li>

</ul>

<p>
La incorporación de estas medidas permitirá fortalecer la resiliencia del sistema de abastecimiento frente a posibles cambios en las condiciones hidrológicas de la cuenca durante el horizonte de diseño del proyecto.
</p>

<p><b>Riesgo 1 — Contaminación de acuíferos por Drenaje Ácido de Mina (DAM):</b></p>

<p>
La infiltración de efluentes ácidos desde las instalaciones mineras puede alterar la calidad fisicoquímica de los manantiales. Los parámetros de mayor riesgo son: pH (acidificación), metales pesados (As, Pb, Cd, Fe, Mn) y sulfatos.
</p>

<p>
Medidas:
establecer línea base de calidad de agua en ambos manantiales antes del inicio de operaciones mineras; monitoreo semestral de metales pesados, pH y conductividad; convenio de monitoreo participativo entre Municipalidad, empresa minera y ANA; y provisión en el diseño del sistema de espacio físico para instalar en el futuro una unidad de tratamiento de agua (filtración lenta + filtración por carbón activado) si los parámetros exceden los LMP del D.S. 031-2010-SA.
</p>

<p><b>Riesgo 2 — Incremento de demanda por crecimiento poblacional inducido:</b></p>

<p>
El inicio de operaciones mineras generará migración laboral y crecimiento de actividad económica en Cascas. Este efecto está parcialmente recogido en el factor de mayoración censal 1.05. Se recomienda una revisión del caudal de diseño a los 5 años de operación del sistema, cuando el impacto del crecimiento minero-inducido sea cuantificable.
</p>

<p><b>Riesgo 3 — Daño físico a la infraestructura por tránsito de vehículos pesados:</b></p>

<p>
El acceso a las instalaciones mineras puede requerir el uso de vías que crucen el trazado de las líneas de conducción. Medidas: coordinar con la empresa minera el trazado de sus vías de acceso para evitar cruce con tuberías; en caso de cruce inevitable, proteger la tubería con losa de concreto armado de 0.15 m de espesor y señalizar adecuadamente.
</p>

<p><b>1.6.6 Lluvias intensas</b></p>

<p>
La temporada de lluvias (diciembre–abril) puede generar caudales pluviales que ingresen al sistema de alcantarillado sanitario por conexiones erradas o por sumideros, incrementando el caudal de llegada a la PTAR. El diseño de la cámara de rejas y del Tanque Imhoff considera un factor de punta hidráulico (<i>Q</i><sub>mh</sub>/<i>Q</i><sub>p</sub> = 2.0) que absorbe estas variaciones. Las tapas de buzones deben ser herméticas para evitar infiltraciones de aguas pluviales.
</p>

<p><b>2.2 Especificaciones Técnicas de las Tuberías</b></p>

<p style="text-align: justify;">El presente acápite define los materiales, clases de presión, dimensiones, condiciones de instalación y pruebas de las tuberías que conforman los sistemas de agua potable y alcantarillado del proyecto. <ul style="text-align: justify;">

  <li><b>NTP ISO 4435</b> — Tubos y conexiones de PVC-U para sistemas de drenaje y alcantarillado sin presión.</li>
  <li><b>Norma Técnica para el Diseño de Opciones Tecnológicas de Saneamiento (NTSS)</b> del sector saneamiento.</li>
</ul>

Las redes de distribución se ejecutan íntegramente en PVC-U de unión flexible (UF) clase 7.5.</p>

La presión disponible en el punto de empalme debe garantizar una presión mínima de servicio de 10 mca.</p>

<p><b>2.2.5 Especificaciones de instalación</b></p>

<ul style="text-align: justify;">
  <li>Excavación de zanjas con ancho y profundidad según diámetro, garantizando un recubrimiento mínimo de 1.00 m sobre la clave del tubo en toda la longitud, como medida frente a huaicos y cargas de tránsito.</li>
  <li>Cama de apoyo de material seleccionado (arena o material fino exento de piedras) de espesor mínimo 0.10 m, conformada y nivelada según la pendiente de diseño.</li>
  <li>Relleno lateral y sobre la clave con material seleccionado compactado por capas; relleno superior con material propio compactado al grado especificado.</li>
  <li>Protección con enrocado y concreto ciclópeo en los cruces de quebrada y zonas susceptibles a erosión; losa de concreto armado en los cruces con vías.</li>
  <li>Las uniones flexibles (espiga-campana con anillo elastomérico) deben instalarse limpias y lubricadas según indicación del fabricante.</li>
</ul>

<ul style="text-align: justify;">
  <li>Cámaras Rompe-Presión Tipo 06, intercaladas en la conducción donde la presión estática supere los 50 mca, a fin de mantener las presiones dentro de la clase de tubería.</li>
  <li>Válvulas de aire (ventosas) en todos los puntos altos del perfil y válvulas de purga en todos los puntos bajos.</li>
  <li>Válvulas de seccionamiento cada 500 m para aislar tramos en caso de rotura o mantenimiento.</li>
</ul>

<p><b>2.2.7 Clases de presión y pruebas hidráulicas</b></p>

<p style="text-align: justify;">Cada clase de tubería admite una presión de trabajo máxima que no debe excederse en condición estática. En obra se ejecuta la prueba hidráulica de presión por tramos y, en el sistema de alcantarillado, la prueba de hermeticidad; previo a la puesta en servicio se realiza la desinfección de las tuberías de agua potable.</p>
